```python
import os
import csv
import time
import datetime

def organize(input_dir,output_dir):
    row_list=[]
    for root, dirs, files in os.walk(input_dir, topdown=False):
        for name in files:
            my_path=os.path.join(root, name)
            file_stats=os.stat(my_path)
            size=file_stats.st_size
            ct = os.path.getctime(my_path)            
            ctime_str = datetime.datetime.fromtimestamp(ct).strftime('%d-%m-%Y %H:%M:%S')
            mt = os.path.getmtime(my_path)
            mtime_str = datetime.datetime.fromtimestamp(mt).strftime('%d-%m-%Y %H:%M:%S')
            at = os.path.getatime(my_path)
            atime_str = datetime.datetime.fromtimestamp(at).strftime('%d-%m-%Y %H:%M:%S')
            type=os.path.splitext(my_path)[-1]
            row_list.append([name, my_path,size,ctime_str,mtime_str,atime_str,type])
    os.makedirs(output_dir, exist_ok=True)
    with open('output.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        field = ["File name" ,"File path", "size","create time","modification time","access time","type" ]
        writer.writerow(field)
        writer.writerows(row_list)
        
```
